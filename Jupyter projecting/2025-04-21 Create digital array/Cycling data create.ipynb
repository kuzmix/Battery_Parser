Конфигурируем способ открытия папок с экспериментами.


```python

def format_experiment(experiment, columns_renaming=None):
    if not columns_renaming:
        columns_renaming = {'Total Time': 't, с', 'I': "I, А", 'E': 'U, В', 'Q': 'Q, Кл'}
    selection_1 = experiment[columns_renaming.keys()]
    selection_1.rename(columns=columns_renaming, inplace=True)
    return selection_1
```

```python
data_dir = r"D:\!Science\Физтех\Циклирования\Результаты-2\Данные"
cycling_n = ['3', '5', '7', '9', '11', ]
cycling_files = defaultdict(list)
for n in cycling_n:
    [cycling_files[os.path.basename(i).split('.')[0]].append(i) for i in
     bp.importing.list_files(os.path.join(data_dir, n), 'csv')]

save_dir = os.path.join(root_dir, 'digital_array')
for pouch, experiments in cycling_files.items():
    for i, filepath in enumerate(experiments, 1):
        data = bp.exporting.load_experiment(filepath)
        result = format_experiment(data)
        bp.exporting.save_experiment(result, os.path.join(save_dir, pouch, f'Этап {i}.csv'), index=False)
```

```python
files = bp.importing.list_files(save_dir, 'csv')
test = Path(files[0])
test.rename(f'Этап {test.stem}')
```

```python
data_dir = r'D:\!Science\Физтех\Циклирования\Результаты-2\Цифровой массив\Циклирования'
list_files = bp.importing.list_files(data_dir, 'csv')
column_renaming = {'t, с': 'time, s', "I, А": 'Current, A', 'U, В': 'Voltage, V', 'Q, Кл': 'Electric charge, C'}
for file in list_files:
    data = bp.exporting.load_experiment(file)
    result = format_experiment(data, column_renaming)
    bp.exporting.save_experiment(result, file, index=False)
```
